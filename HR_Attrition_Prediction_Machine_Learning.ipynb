```python
import pandas as pd

# Western  Forest Service

# upload data from csv
forest_1_df = pd.read_csv('C:\DAT-223\\ForestCoverTypeDatasetSubset_1.csv')
forest_2_df = pd.read_csv('C:\DAT-223\\ForestCoverTypeDatasetSubset_2.csv')

# merge the two datasets
forest_merged = forest_1_df.merge(forest_2_df, left_on='key', right_on='key')

# fill all missing values with 0s
forest_merged.fillna(0, inplace=True)

# double check to ensure merge was done properly
print(f'original dataset sizes:  forest_1_df={forest_1_df.shape},  forest_2_df={forest_2_df.shape}' )
print(f'merged dataset size:  forest_merged={forest_merged.shape}')

# print the first 10 rows of data
print(forest_merged.head(10))
```
